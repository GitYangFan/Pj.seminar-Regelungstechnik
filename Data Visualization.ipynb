### according to the figures, i guess the orientations of axes look like following:
### x right
### y down
### z forward

reason: (assume that all data from imu use the same coordinate system, and the car move 'forward')

if the orientation of the following 2D curve figure is correct, the car turned left at the beginning, which corresponds the negative \
angular velocity of axis y at the beginning(imu). So the orientation of y is downward.

In the figure of 'velocity', we can note that the car was accelerated to a constant velocity, which corresponds the linear acceleration z.\
It has a positive peak at the beginning and keep arround 0 after that. So the orientation of z is forward, which determinds: x right.

about the gravity: about -10m/s along axis y, which means its direction is upward. So this value is actually not the acceleration of \
gravity, but the acceleration of the car. The effect of gravity is similar to the situation: a force lift the car towards the sky.

```python
from get_data import *

import numpy as np
from bokeh.io import output_notebook, show
from bokeh.layouts import gridplot
from bokeh.plotting import figure
output_notebook()

def plot_config(width: int, height: int, title=None, x_label: str = None, y_label: str = None):
    p = figure(title=title, background_fill_color="#fafafa", width=width, height=height)
    p.output_backend = "svg"  # save plot in HTML as SVG
    p.title.text_font_size = "15pt"
    p.xaxis.axis_label = x_label
    p.yaxis.axis_label = y_label
    p.yaxis.axis_label_text_font_size = "12pt"
    p.yaxis.major_label_text_font_size = "12pt"
    p.xaxis.major_label_text_font_size = "12pt"
    p.xaxis.axis_label_text_font_size = "12pt"
    return p
```

```python
global_start_time = df_raw.iloc[0,0]

t_imu = imu['time'] - global_start_time # start from global time 0s
t_tf = tf['time'] - global_start_time
t_vel = vel['time'] - global_start_time
```
